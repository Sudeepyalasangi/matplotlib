```python
x =[3,1,2,5]
y =[5,2,4,7]
```

```python
plt.plot(x, y, 'b*')
for i, j in zip(x, y):
    plt.text(i, j+0.3, '({},{})'.format(i, j))
plt.show()
```

```python
x =[3,1,2,5]
y =[5,2,4,7]
```

```python
plt.plot(x, y, 'b*')
plt.axis([0, 6, 0, 20])
for i, j in zip(x, y):
    plt.text(i, j+0.9, '({},{})'.format(i, j))
plt.show()
```
